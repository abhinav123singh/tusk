```python
#Assigning k to 5
k=5

centroids = {
    i+1: [np.random.randint(0, 100), np.random.randint(0, 150), np.random.randint(0, 100)]
    for i in range(k)
}

```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age, df["Annual Income (k$)"], df["Spending Score (1-100)"], color='k')

ax.view_init(35, 35)
colmap = {1: 'coral', 2: 'g', 3: 'b', 4: 'r', 5: 'y'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

```python
## Assignment Stage

def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1-x2)^2 - (y1-y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df.Age - centroids[i][0]) ** 2
                + (df['Annual Income (k$)'] - centroids[i][1]) ** 2
                + (df['Spending Score (1-100)'] - centroids[i][2]) ** 2
            )
        )
    centroids_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroids_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)
print(df)
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age, df["Annual Income (k$)"], df["Spending Score (1-100)"], color=df['color'], alpha=0.5, edgecolor='k')

ax.view_init(30, 185)
colmap = {1: 'coral', 2: 'g', 3: 'b', 4: 'r', 5: 'y'}
for i in centroids.keys():
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel('Spending Score (1-100)')
plt.show()
```

```python
#Update Stage

import copy

old_centroids = copy.deepcopy(centroids)

def update(k):
    for i in centroids.keys():
        centroids[i][0] = np.mean(df[df['closest'] == i]["Age"])
        centroids[i][1] = np.mean(df[df['closest'] == i]["Annual Income (k$)"])
        centroids[i][2] = np.mean(df[df['closest'] == i]["Spending Score (1-100)"])
    return k

centroids = update(centroids)

centroids
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df.Age, df["Annual Income (k$)"], df["Spending Score (1-100)"], color=df['color'], alpha=0.5, edgecolor='k')

ax.view_init(30, 185)
colmap = {1: 'coral', 2: 'g', 3: 'b', 4: 'r', 5: 'y'}
for i in centroids.keys():
    old_x = old_centroids[i][0]
    old_y = old_centroids[i][1]
    old_z = old_centroids[i][2]
    dx = (centroids[i][0] - old_centroids[i][0])*0.75
    dy = (centroids[i][1] - old_centroids[i][1])*0.75
    dz = (centroids[i][2] - old_centroids[i][2])*0.75
    ax.arrow(old_x, old_y, old_z, dx, dy, dz, head_width=2, head_length=3, fc=colmap[i], ec=colmap[i])
    plt.scatter(*centroids[i], color=colmap[i])
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
ax.set_zlabel("Spending Score (1-100)")
plt.show()
```

```python
## Assignment Stage

def assignment(df, centroids):
    for i in centroids.keys():
        # sqrt((x1-x2)^2 - (y1-y2)^2)
        df['distance_from_{}'.format(i)] = (
            np.sqrt(
                (df.Age - centroids[i][0]) ** 2
                + (df['Annual Income (k$)'] - centroids[i][1]) ** 2
                + (df['Spending Score (1-100)'] - centroids[i][2]) ** 2
            )
        )
    centroids_distance_cols = ['distance_from_{}'.format(i) for i in centroids.keys()]
    df['closest'] = df.loc[:, centroids_distance_cols].idxmin(axis=1)
    df['closest'] = df['closest'].map(lambda x: int(x.lstrip('distance_from_')))
    df['color'] = df['closest'].map(lambda x: colmap[x])
    return df

df = assignment(df, centroids)
print(df)
```

```python
#Continue until all assignment categories don't change anymore
while True:
    closest_centroids = df['closest'].copy(deep=True)
    centroids = update(centroids)
    df = assignment(df, centroids)
    if closest_centroids.equals(df['closest']):
        break
```
